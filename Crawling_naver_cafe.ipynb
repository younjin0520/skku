```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import openpyxl #엑셀사용하기
import time

driver = webdriver.Chrome('./chromedriver.exe')
wb = openpyxl.Workbook()
sheet = wb.active #sheet활성화
sheet.append(["검색어","태블릿명","가격","등급","날짜"])
search = input("검색어 입력: ")
num=0
p=0
for p in range(1,91):
    #url에서 데이터 가져오기
    driver.implicitly_wait(3)#로딩을 위해 3초 기다림
    driver.get("https://cafe.naver.com/ca-fe/home/search/c-articles?q={name}&p={page}&ss=COMPLETED".format(name = search+"중고",page=p))
    time.sleep(1)
    #HTML코드 파싱
    html = BeautifulSoup(driver.page_source, "html.parser")
    #select함수는 컨테이너를 차례로 리스트로 저장
    container = html.select("div.detail_area")
    for con in container:
        num=0
        name2 = con.select_one("span.subject_text")
        date = con.select_one("span.write_date_time")
        for span in con.find_all("span", class_="info_text"):
            if(num==0):
                price=span.text
            elif(num==2):
                state=span.text
            elif(num>3):
                num=0
                continue
            num=num+1
        print(price,state)
        sheet.append([search,name2.text, price,state,date.text])
    print(p)
wb.save("tablet_"+search+".xlsx")
```
