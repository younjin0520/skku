```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import openpyxl #엑셀사용하기
import urllib.request
import time
import os
```

```python
#폴더생성
os.makedirs('./new_image')
```

```python

driver = webdriver.Chrome('./chromedriver.exe')
wb = openpyxl.Workbook()
sheet = wb.active #sheet활성화
sheet.append(["image","state"])
search = input("검색어 입력: ")
num=0
name=1
p=0

for p in range(1,91):
    #url에서 데이터 가져오기
    url = "https://cafe.naver.com/ca-fe/home/search/c-articles?q={name}&p={page}&ss=COMPLETED".format(name = search+"중고",page=p)
    driver.implicitly_wait(3)#로딩을 위해 3초 기다림
    driver.get(url)
    time.sleep(1)
    #HTML코드 파싱
    html = BeautifulSoup(driver.page_source, "html.parser")
    #select함수는 컨테이너를 차례로 리스트로 저장
    container = html.select("div.item_inner")
    try:
        for con in container:
            num=0
            soup_img=con.find("a",class_="thumbnail_area")
            img = soup_img.select_one('img')['src']
            urllib.request.urlretrieve(img,"./new_image/"+search+"_"+str(name)+".jpg")
        
            for span in con.find_all("span", class_="info_text"):
                if(num==2):
                    state=span.text
                elif(num>3):
                    num=0
                    continue
                num=num+1
            print(name,state)
            sheet.append([search+"_"+str(name),state])
            name=name+1
    except:
        continue
    print(p)
wb.save("image_"+search+".xlsx")
```
