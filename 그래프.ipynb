```python
import pandas as pd
import numpy as np

#파일 불러오기
data = pd.read_excel('./data_new.xlsx')
data
```

```python
import math
import matplotlib.pyplot as plt
#변수와의 상관관계 바이올린 플롯으로 그리기관관계 바이올린 플롯으로 그리기
import seaborn as sns
plt.xlabel('state')
plt.ylabel('price')
sns.violinplot(x=data['state'],y=data['price'],data=data)
plt.show()
```

```python
#변수와의 상관관계
plt.xlabel('price_i')
plt.ylabel('price')
plt.scatter(data[['price_i']],data[['price']],alpha=0.1,color="gray")
plt.show()
```

```python
#변수와의 상관관계
plt.xlabel('used')
plt.ylabel('price')
plt.scatter(data[['used']],data[['price']],alpha=0.1,color="gray")
plt.show()
```
